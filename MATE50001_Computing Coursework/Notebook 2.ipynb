```python
import M

model = M.Model('healthcare-dataset-stroke-data.csv', 'stroke')
model.get_data()
```

```python
# cv_f1-score and f1-score with whole var
cvf1, score = model.with_wholevar(['LogisticRegression', 'KNeighborsClassifier', 'SVC'])
cvf1
```

```python
# presenting the cross-validation score from different models
# click the blue button to change the model
model.draw_wholedata(cvf1)
```

The graph above shows the f1-score for different model with different parameters.

For Logistic Regression with class_weight = 'balanced', we change C parameter ---> C: [0.5,1.0,1.5] = parameters: [0,1,2] in the graph.

For KNeighborsClassifier with weights = 'distance', we change n_neighbors parameter ---> n_neighbors: [2,3,4,5] = parameters: [2,3,4,5] in the graph.

For SVC with class_weight = 'balanced', we change kernels parameter ---> kernels: ['linear', 'poly', 'rbf'] = parameters: [0,1,2] in the graph.

We choose class_weight = 'balanced'and weights = 'distance' just because we have much more 0 than 1.

We can see that Logestic regression with balanced class_weight has the highest f1-score for all three C values. So we choose Logestic regression with balanced class_weight to be the final model and in notebook3 we can see how the f1-score change for the model if we delete one column of variable.
