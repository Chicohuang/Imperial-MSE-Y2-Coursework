```python
import M
import sklearn.metrics
import pandas as pd
from sklearn.linear_model import LogisticRegression

model = M.Model('healthcare-dataset-stroke-data.csv', 'stroke')
model.get_data()
```

```python
# presenting the data after processing
model.x_data
```

```python
x_train, x_test, y_train, y_test = model.split_data(model.x_data.to_numpy(),model.y_data.to_numpy())
```

```python
# simply use Logestic Regression 
lr = LogisticRegression(max_iter=3000).fit(x_train, y_train)
y_pred = lr.predict(x_test)

# produce confusion matrix of the model
pd.DataFrame(sklearn.metrics.confusion_matrix(y_test, y_pred))
```

From the confusion matrix we can see that the model predict that all patients don't have stroke.

Because there are much more patients without stroke in real data, so although all predict values are 0,the model has quite good accuracy_score(0.957). However it has very bad precision and recall score(0.0), therefore we have to choose f1_score instead of accuracy to see whether the model is good.

Unfortunately f1 score is quite low (0.0), so it means Logistic Regression with default parameter is not a good choice, so in notebook2 we will choose different models with different parameters to see which one has the largest f1 score and can be used to predict whether a patient has stroke.
